```python
import os
import json
```

```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
con = sqlite3.connect("database.sqlite")
```

```python
# select emails sent to hillary clinton
df = pd.read_sql("select * from Emails e left join EmailReceivers r on e.Id = r.EmailId where r.PersonId=80", con)
```

```python
export = []
for i, r in df.iterrows():
    if r['MetadataSubject'].strip() == "":
        subj = "<b>Subject:</b> none<br/>"
    else:
        subj = f"<b>Subject:</b> <i>{r['MetadataSubject'].strip()}</i><br/>"
    if r['ExtractedBodyText'].strip() == "":
        body = "<b>Body</b>: none<br/>"
    else:
        body = f"<b>Body</b>:<pre class='summary'>{r['ExtractedBodyText'].strip()}</pre><br/>"
    txt = f"{subj}{body}<b>RawText</b>:<pre>{r['RawText']}</pre>"
    export.append({"id": str(r['EmailId']), "data": {"text": txt, "docNumber": r['DocNumber'], "dbId": str(r['EmailId'])}})
```

```python
output = json.dumps(export, indent=2)
with open('labelstudio_export.json', 'w') as f:
    f.write(output)
```
